```python
def sort_test_directory(dir):
    df1 = pd.read_csv(os.path.join(dir, 'model_log.csv'), na_values='None')
    df1 = df1[df1.fold_str!='sum']
    return df1.groupby('identifier').mean().sort_values('valid_gini', ascending=False)
```

```python
sort_test_directory('../../model/raw_data/')
```
