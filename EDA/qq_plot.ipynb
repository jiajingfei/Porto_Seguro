```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.graphics.gofplots as sm
```

```python
train = pd.read_csv('../Kaggle1/train.csv')
```

```python
pos_sample = train[train.target == 1]
neg_sample = train[train.target == 0].sample(n=pos_sample.shape[0])
```

```python
for c in train.columns:
    if (not c.endswith(('bin')) and c.startswith('ps')):
        print c
        _ = sm.qqplot_2samples(pos_sample[c], neg_sample[c], line='45')
        plt.show()
        
# feature with significant different distribution 
# ps_ind_05_cat
# ps_reg_02
# ps_reg_03
# ps_car_04_cat
# ps_car_12
# ps_car_13
# ps_car_14
# ps_car_15
# ps_calc_13 tail
# ps_calc_14 tail
```
